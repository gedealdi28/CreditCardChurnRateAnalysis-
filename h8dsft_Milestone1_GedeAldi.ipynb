```python
###  Visualisasi Data untuk penjabaran masalah pertama 
'''Penjabaran 1 yaitu Bagaimana gambaran umum demografi dari Customer dijawab dengan membuat
pie chart persen distribusi dari kelompok umur, gender, level edukasi, dan status pernikahannya
'''  
df['Age_Group'] = pd.cut(df['Customer_Age'], bins=range(20, 90, 10), labels=[f'{i}-{i+9}' for i in range(20, 80, 10)], right=False)
## Gambaran Umum customer dari Umur, Gender, Education dan Marital Status
total_by_age = df.groupby('Age_Group').size().reset_index(name='AgeGroup_Count')
total_by_gender=df.groupby('Gender').size().reset_index(name='TotalGender')
total_by_education=df.groupby('Education_Level').size().reset_index(name='TotalEducation')
total_by_marital=df.groupby('Marital_Status').size().reset_index(name='TotalMarital')


fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 18))
## Barplot
axes[0,0].pie(total_by_age['AgeGroup_Count'],labels= total_by_age['Age_Group'],autopct='%1.1f%%', startangle=90)
axes[0,0].set_title('Total Pelanggan Berdasarkan Usia')

axes[0,1].pie(total_by_gender['TotalGender'],labels= total_by_gender['Gender'],autopct='%1.1f%%', startangle=90)
axes[0,1].set_title('Total Pelanggan Berdasarkan Gender')

axes[1,0].pie(total_by_education['TotalEducation'],labels= total_by_education['Education_Level'],autopct='%1.1f%%', startangle=90)
axes[1,0].set_title('Total Pelanggan Berdasarkan Level Pendidikan')

axes[1,1].pie(total_by_marital['TotalMarital'],labels= total_by_marital['Marital_Status'],autopct='%1.1f%%', startangle=90)
axes[1,1].set_title('Total Pelanggan Berdasarkan Marital Status')



plt.tight_layout()
plt.show()
```

```python
#### Apakah terdapat perbedaan signifikan dalam tingkat churn antara jenis kartu
###H0 (Hipotesis Nol): Tidak ada perbedaan signifikan dalam rata-rata Limit Kredit antara jenis kartu kredit (Blue, Gold, Silver, Platinum).
###H1 (Hipotesis 1): Terdapat perbedaan signifikan dalam rata-rata Limit Kredit antara setidaknya jenis kartu kredit.
from scipy import stats
from scipy.stats import f_oneway

blue = df['Credit_Limit'][(df['Card_Category'] == 'Blue')]
gold = df['Credit_Limit'][(df['Card_Category'] == 'Gold')]
silver = df['Credit_Limit'][(df['Card_Category'] == 'Silver')]
platinum = df['Credit_Limit'][(df['Card_Category'] == 'Platinum')]

res = stats.f_oneway(blue, gold, silver, platinum)

print('P-value:', res.pvalue)
                           
```

Dilakukan pengujian Anova untuk melihat perbedaan yang signifikan antara rata-rata dari kredit limit dan jenis kartu kredit. Di mana hipotesis nolnya adalah tidak ada perbedaan signifikan dalam rata-rata kredit limit antara jenis kartu kredit dan hipotesis alternatifnya terdapat perbedaan signifikan dalam rata-rata kredit limit antara setidaknya dua jenis kartu kredit. Dapat dilihat bahwa nilai P value itu bernilai nol dan kurang dari 0,05 ini menandakan bahwa kita dapat menolak h0 dan menerima hipotesis alternatif yaitu terdapat perbedaan signifikan dalam rata-rata kredit limit antara jenis kartu kredit.
